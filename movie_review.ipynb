```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec= CountVectorizer(min_df=1,tokenizer= nltk.word_tokenize)
```

```python
sents= df['Phrase']
len(sents)
```

```python
sent_count=vec.fit(sents)
sentcount=vec.transform(sents)
```

```python
sent_count
```

```python
vec.vocabulary_
```

```python
sentcount.shape
```

```python
sent_count.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()
sent_tfidf = tfidf_transformer.fit_transform(sentcount)
```

```python
sent_tfidf
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

x= df.drop(['Sentiment'])
y= df['Sentiment']

```python
y= df['Sentiment']
```

```python
x_train,x_test,y_train,y_test=train_test_split(sent_tfidf,y,test_size=0.25)
```

```python
clf = MultinomialNB()
clf.fit(x_train,y_train)
```

```python
y_pred= clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
#classification_report(y_test,y_pred)
```
